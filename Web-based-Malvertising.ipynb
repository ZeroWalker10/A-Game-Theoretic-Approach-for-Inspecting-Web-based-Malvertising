```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

# simulation 1

```python
alpha = np.arange(0.00, 1.05, 0.05)
Ci = 0.4
Cm = 0.3
g = 0.9
ls = np.array([0.6, 0.7, 0.8])
```

```python
marker = ['D', 's', '^']
color = ['blue', 'red', 'limegreen']
y_ticks = np.linspace(-0.9, 0.0, 10)
for l, m, c in zip(ls, marker, color):
    defender_payoffs = []
    for a in alpha:
        if a <= Ci / l:
            defender_payoffs.append(-l)
        elif a <= (g - Cm) / g:
            defender_payoffs.append(-Ci - (1 - a) * l)
        else:
            y = (g - Cm) / (a * g)
            x = (a * l - Ci) / (a * l)
            I_payoff = (-Ci - (1 - a) * l) * y * (1 - x) + (-Ci) * y * x
            NoI_payoff = 0 * (1 - y) * x + (-l) * (1 - y) * (1 - x)
            defender_payoffs.append(I_payoff + NoI_payoff)
    plt.ylim([-0.9, 0.0])
    plt.yticks(y_ticks)
    plt.plot(alpha, defender_payoffs, marker=m, color=c, label='l={}'.format(l))
plt.legend()
plt.xlabel(r'Detection rate($\alpha$)')
plt.ylabel('Defender Payoff')
plt.grid(True)
plt.show()
```

# simulation 2

```python
alpha = np.arange(0.00, 1.05, 0.05)
Ci = 0.4
Cm = 0.3
l = 0.7
gs = np.array([0.9, 0.8, 0.7])
```

```python
marker = ['D', 's', '^']
color = ['blue', 'red', 'limegreen']
y_ticks = np.linspace(-0.1, 0.7, 9)
for g, m, c in zip(gs, marker, color):
    attacker_payoffs = []
    for a in alpha:
        if a <= Ci / l:
            attacker_payoffs.append(-Cm + g)
        elif a <= (g - Cm) / g:
            attacker_payoffs.append(-Cm + (1 - a) * g)
        else:
            y = (g - Cm) / (a * g)
            x = (a * l - Ci) / (a * l)
            B_payoff = 0 * x * y + 0 * x * (1 - y)
            M_payoff = (-Cm + (1 - a) * g) * (1 - x) * y + (-Cm + g) * (1 - x) * (1 - y)
            attacker_payoffs.append(B_payoff + M_payoff)
    plt.ylim([-0.1, 0.7])
    plt.yticks(y_ticks)
    plt.plot(alpha, attacker_payoffs, marker=m, color=c, label='g={}'.format(g))
plt.grid(True)
plt.xlabel(r'Detection rate($\alpha$)')
plt.ylabel('Attacker Payoff')
plt.legend()
plt.show()
```
